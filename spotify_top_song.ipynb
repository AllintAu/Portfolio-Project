# Preparation

```python
# Create new column to store artist name list, in case there were more than 1 artist contributed
# Create new column to store only main artist name
df['artist_list'] = df.artist_names.str.split(",")
df['first_artist'] = df['artist_list'].apply(lambda x: x[0].strip() if x else None)
```

```python
# Convert Data Type
df['streams'] = df['streams'].astype(float)
df['weeks_on_chart'] = df['weeks_on_chart'].astype(float)
```

# Artists With Most Week on Chart

First Case Weeks on Chart

```python
# In 1 track there might have more than 1 artist. And the name of artist of that track most likely
# start with main artist followed by feturing artist. 
# This first case consider number of Week On Chart to every artist that contributed to that track equaly 

artist_dic = {}

def applyAllArtist(record):
    global artist_dic
    
    for artist in record['artist_list']:
        artist = artist.strip()
        # Check did specific artist already exists in dictionary
        if artist in artist_dic:
            artist_dic[artist] += record["weeks_on_chart"]               
        else:
            artist_dic[artist] = record["weeks_on_chart"]
                
                
df.apply(applyAllArtist, axis = 1)

print("Funtion applyAllArtist Applied")
```

```python
df_all_chartweek = pd.DataFrame(artist_dic, index = [0]).T.sort_values(0, ascending = False)
df_all_chartweek.rename(columns = {0:"Total Week on Chart"}, inplace = True)
df_all_chartweek.head()
```

Second Case Weeks on Chart

```python
# This case will give Main Artist wiht full of Week On Chart value
# For Featuring Artist that contributed to the track will get half of the Week On Chart value

artist_dic_full_n_half = {}

def applyFullAndHalf(record):
    global artist_dic_full_n_half
    num_artists = len(record['artist_list'])
    
    for index, artist in enumerate(record['artist_list']):
        artist = artist.strip()
        # Check did specific artist already exists in dictionary
        
        if artist in artist_dic_full_n_half:
            if index == 0: # first artist gets full week_on_chart value
                artist_dic_full_n_half[artist] += record["weeks_on_chart"]
                
            else: # other artist get half of week_on_chart value
                artist_dic_full_n_half[artist] += record["weeks_on_chart"] / 2
                
        else:
            if index == 0: # First artist gets full weeks_on_chart values
                artist_dic_full_n_half[artist] = record["weeks_on_chart"]
                
            else:
                artist_dic_full_n_half[artist] = record["weeks_on_chart"] / 2 
                
                
df.apply(applyFullAndHalf, axis = 1)

print("Funtion applyFullAndHalf Applied")
```

```python
df_full_n_half_chartweek = pd.DataFrame(artist_dic_full_n_half, index = [0]).T.sort_values(0, ascending = False)
df_full_n_half_chartweek.rename(columns = {0:"Total Week on Chart"}, inplace = True)
df_full_n_half_chartweek.head()
```

Third Case Weeks on Chart

```python
# In this case we only give the Week On Chart value to the Main Artist for measuring individual artist or band

first_artist_only_dic = {}

def applyFirstOnly(record):
    global first_artist_only_dic
    
    # only process the first artist in the list
    if record['artist_list']: # check is list empty 
        first_artist = record['artist_list'][0].strip()  # Locate the first artist in the list
        
        if first_artist in first_artist_only_dic:
            first_artist_only_dic[first_artist] += record['weeks_on_chart']
            
        else:
            first_artist_only_dic[first_artist] = record['weeks_on_chart']
            
df.apply(applyFirstOnly, axis = 1)

print("Function applyFirstOnly Applied")
```

```python
df_first_only_chartweek = pd.DataFrame(first_artist_only_dic, index = [0]).T.sort_values(0, ascending = False)
df_first_only_chartweek.rename(columns = {0:"Total Week on Chart"}, inplace = True)
df_first_only_chartweek.head()
```

Alternative Way to process Third Case Weeks on Chart

```python
# Another way to do third case
# Extract main artist name from artist_list 
# Groupby artist name then sum up weeks_on_chart 

group_main_artist = df.groupby(by='first_artist')

group_week_sum = group_main_artist['weeks_on_chart'].sum()

group_week_sum.sort_values(ascending = False).head()
```
